```python
# grabbing specific item not returning a number -> tensor(number)
tsn = torch.arange(10)
# tsn[7] #  tensor(7)
tsn = tsn.reshape(-1, 2)
tsn
```

```python
tsn2 = torch.arange(10)
tsn2 = tsn2.reshape(-1, 2)
tsn2 = tsn2[:, 1:] # slicing
# u don't have to do so cause it was already as a vector before slicing
# tsn = tsn.reshape(-1, 1)
tsn2
```
